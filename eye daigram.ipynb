```python
import numpy as np
import matplotlib.pyplot as plt
```

random bit generation

```python
bits = np.random.randint(0, 2,20)
print(bits)
bits=[-1 if x==0 else 1 for x in bits]
print(bits)
```

samples per symbols

```python
sps=8
new_bits = []
for i in bits:
    new_bits.append(i)
    for j in range(sps):
        new_bits.append(0)
print(new_bits)
```

Raised Cosine

```python

Ts = 10
beta=0.75
t = np.arange(-50, 51)
h = 1/Ts * np.sinc(t/Ts) * np.cos(np.pi*beta*t/Ts) / (1 - (2*beta*t/Ts)**2)

# Plot the filter
plt.figure(1)
plt.plot(t, h)
plt.grid(True)
plt.show()
```

*convole the signal*

```python
y=np.convolve(bits,h)
plt.plot(y)
plt.show()
```

```python
T=1
Fs=100
x = np.arange(-T, T, 1 / Fs)
for i in range(2 * Fs, len(y) - 3 * Fs, Fs):
    plt.plot(x, y[i:i + 2 * Fs], 'blue')
```
