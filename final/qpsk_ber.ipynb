```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N=50000
EbNo_db=np.arange(0,50)
ber=[]
for i in EbNo_db:
    Ebno=10**(i/10)
    noise=1/np.sqrt(2*Ebno)
    mes=np.random.rand(N)>=0.5
    mess=[str(int(i)) for i in mes]
    mess="".join(mess)
    message=np.array([mess[2*i:2*(i+1)] for i in range(int(len(mes)/2))])
    r=mes+np.random.randn(N)*noise>=0.5
    rec=[str(int(i)) for i in r]
    rec="".join(rec)
    received=np.array([rec[2*i:2*(i+1)] for i in range(int(len(mes)/2))])
    error=(message!=received).sum()
    ber.append(error/N)
plt.semilogy(EbNo_db,ber,"*-")
plt.xlabel('Eb/No')
plt.ylabel('BER')
plt.title('QPSK')
    
```

```python
import numpy as np

N = 50000
EbNo_db = np.arange(0, 50)
ber = []

for i in EbNo_db:
    EbNo = 10**(i / 10)
    noise_std_dev = 1 / np.sqrt(2 * EbNo)
    
    # Generate random binary message
    message = np.random.randint(0, 2, N)
    
    # Simulate transmission over noisy channel
    noise = noise_std_dev * np.random.randn(N)
    received_signal = message + noise
    
    # Apply threshold to received signal to get received bits
    received = received_signal >= 0.5
    
    # Calculate bit errors
    errors = (message != received).sum()
    
    # Calculate BER and append to list
    ber.append(errors / N)

# Print the BER for inspection
print(ber)
plt.plot(EbNo_db, ber, '*-')
plt.xlabel('Eb/No')
plt.ylabel('BER')
plt.title('QPSK')
```
