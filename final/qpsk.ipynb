```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fc=30
fs=30*fc
ts=np.arange(0,2/fc,1/fs)
print(len(ts))
```

```python
decimal=input("enter the roll number:")
binary=[format(int(i),'04b') for i in decimal]
bcd="".join(binary)
print(bcd)
```

```python
carrier={
    "00":np.cos(2*np.pi*fc*ts+3*np.pi/4),
    "01":np.cos(2*np.pi*fc*ts+np.pi/4),
    "10":np.cos(2*np.pi*fc*ts-3*np.pi/4),
    "11":np.cos(2*np.pi*fc*ts-np.pi/4)
}
```

```python
# bcd="00011110"
joined=[bcd[i:i+2] for i in range(0,len(bcd),2)]
print(joined)
```

```python
qpsk=np.concatenate([carrier[i] for i in joined])
plt.plot(np.arange(0,len(qpsk)),qpsk)
```
